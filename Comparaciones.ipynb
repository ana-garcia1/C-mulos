```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ('Alessi 18',
'Archinal 1',
'Berkeley 82',
'ESO 312-04',
'ESO 430-09',
'ESO 435-48',
'ESO 445-74',
'ESO 518-03',
'NGC 6204',
'vdBH 92')
```

```python
#mis valores:
Garcia = ( 2.00,
0.20,
0.20,
0.10, 
0.20,
1.00,
3.00,
2.00,
0.23,
0.03)

Garcia_err_inf = (1,
              0.1,
              0.1,
              0.05,
              0.05,
              0.7,
              1,
              0.5,
              0.03,
              0.02)

Garcia_err_sup = (1,
              0.1,
              0.1,
              0.05,
              0.05,
              0.2,
              1,
              0.5,
              0.03,
              0.02)
```

```python
Garcia_yr = []
for i in range(len(Garcia)):
    Garcia_yr.append( Garcia[i] * 1000000000)

Garcia_err_inf_yr = []
for i in range(len(Garcia)):
    Garcia_err_inf_yr.append( Garcia_err_inf[i] * 1000000000)

Garcia_err_sup_yr = []
for i in range(len(Garcia)):
    Garcia_err_sup_yr.append( Garcia_err_sup[i] * 1000000000)
```

```python
Kharchenko = ( 1.00,
0.10, 
0.10, 
0.20,
0.30, 
0.50, 
3.00, 
1.00, 
0.10,
0.80)

Kharchenko_err_inf = (0.20,
                      0.04,
                      0.05,
                      0.04,
                      0.07,
                      0.10,
                      0.70,
                      0.30,
                      0.04,
                      0.20)
                      
Kharchenko_err_sup = (0.30,
                      0.08,
                      0.09,
                      0.05,
                      0.08,
                      0.10,
                      0.80,
                      0.40,
                      0.08,
                      0.20)
```

```python
Kharchenko_yr = []
for i in range(len(Kharchenko)):
    Kharchenko_yr.append( Kharchenko[i] * 1000000000)

Kharchenko_err_inf_yr = []
for i in range(len(Kharchenko)):
    Kharchenko_err_inf_yr.append( Kharchenko_err_inf[i] * 1000000000)

Kharchenko_err_sup_yr = []
for i in range(len(Kharchenko)):
    Kharchenko_err_sup_yr.append( Kharchenko_err_sup[i] * 1000000000)
```

```python
Dias = ( 0.7,
        0.010,
        0.20,
        0.10,
        0,
        0,
        0,
        2,
        0.1,
        0.03)

Dias_err_inf = (0.5,
                0.003,
                0.09,
                0.09,
                0,
                0,
                0,
                0.3,
                0.04,
                0.02)
                
Dias_err_sup = (3,
                0.004,
                0.2,
                0.5,
                0,
                0,
                0,
                0.3,
                0.07,
                0.04)
```

```python
Dias_yr = []
for i in range(len(Dias)):
    Dias_yr.append( Dias[i] * 1000000000)

Dias_err_inf_yr = []
for i in range(len(Dias)):
    Dias_err_inf_yr.append( Dias_err_inf[i] * 1000000000)

Dias_err_sup_yr = []
for i in range(len(Dias)):
    Dias_err_sup_yr.append( Dias_err_sup[i] * 1000000000)
```

```python
CG= ( 0.005,
     0.10,
     0,
     0.05,
     0,
     0,
     0,
     1,
     0.09,
     0.2)

CG_err_inf = (0.002,
              0.04,
              0,
              0.02,
              0,
              0,
              0,
              0.4,
              0.03,
              0.06)

              
CG_err_sup = (0.003,
              0.06,
              0,
              0.030,
              0,
              0,
              0,
              0.6,
              0.06,
              0.09)
```

```python
CG_yr = []
for i in range(len(CG)):
    CG_yr.append(CG[i] * 1000000000)

CG_err_inf_yr = []
for i in range(len(CG)):
    CG_err_inf_yr.append(CG_err_inf[i] * 1000000000)

CG_err_sup_yr = []
for i in range(len(CG)):
    CG_err_sup_yr.append(CG_err_sup[i] * 1000000000)
```

```python
plt.figure(figsize=(15,6))


plt.subplot()

plt.grid(color='gray', linestyle='--', linewidth=0.8, alpha= 0.5)

x = np.arange(0, 10)

plt.scatter(x+0.02, Garcia_yr, marker='*', c='royalblue', s = 200, alpha=0.8, label='Este trabajo')
plt.errorbar(x+0.02, Garcia_yr, yerr=(Garcia_err_inf_yr, Garcia_err_sup_yr), fmt='.', c='royalblue', elinewidth=2, capsize=5, capthick=2, alpha=0.8)

# Obtener el objeto de los ejes actuales
ax = plt.gca()



# Cambiar el color de las etiquetas de los ejes (ticks)
ax.tick_params(axis='x', labelsize=14, rotation=15)  # Color de las etiquetas en el eje X
ax.tick_params(axis='y', labelsize=14)  # Color de las etiquetas en el eje Y


# Cambiar el color de las etiquetas de los ejes (nombres de los ejes X e Y)
ax.set_xlabel('', fontsize=14)
ax.set_ylabel('Nombre del Eje Y')


for i in range(len(labels)):
    
    if Kharchenko[i] != 0:
        plt.scatter(x[i]-0.02, Kharchenko_yr[i], marker='s', c='gold', s = 100, label='Kharchenko')
        plt.errorbar(x[i]-0.02, Kharchenko_yr[i], yerr=[[Kharchenko_err_inf_yr[i]], [Kharchenko_err_sup_yr[i]]], fmt='.', c='gold', elinewidth=2, capsize=5, capthick=2)
        
    if Dias[i] != 0:
        plt.scatter(x[i]+0.07, Dias_yr[i], marker='^', c='orangered', s = 100, alpha=0.8)
        plt.errorbar(x[i]+0.07, Dias_yr[i], yerr=[[Dias_err_inf_yr[i]], [Dias_err_sup_yr[i]]], fmt='.', c='orangered', elinewidth=2, capsize=5, capthick=2, alpha=0.8)

    if CG[i] != 0:
        plt.scatter(x[i]-0.07, CG_yr[i], marker='X', c='limegreen', s = 100, alpha=0.8)
        plt.errorbar(x[i]-0.07, CG_yr[i], yerr=[[CG_err_inf_yr[i]], [CG_err_sup_yr[i]]], fmt='.', c='limegreen', elinewidth=2, capsize=5, capthick=2, alpha=0.8)

plt.xticks(x, labels)
plt.ylabel('Edad [aos]', fontsize=14)
plt.legend(['Este trabajo', 'K13', 'D21', 'CG20'], markerscale=0.7, ncols=2, fontsize=14)

plt.yscale('log')

plt.savefig('comparaciones_articulo2.png', bbox_inches='tight')
```
