```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ('Alessi 18',
'Archinal 1',
'Berkeley 82',
'ESO 312-04',
'ESO 430-09',
'ESO 435-48',
'ESO 445-74',
'ESO 518-SC03',
'BH 92',
'NGC 6204')
```

```python
#mis valores:
Garcia = ( 2.00,
0.20,
0.20,
0.10, 
0.20,
1.00,
3.00,
2.00,
0.03,
0.23)

Garcia_err_inf = (1,
              0.1,
              0.1,
              0.05,
              0.05,
              0.7,
              1,
              0.5,
              0.02,
              0.03)

Garcia_err_sup = (1,
              0.1,
              0.1,
              0.05,
              0.05,
              0.2,
              1,
              0.5,
              0.02,
              0.03)
```

```python
Kharchenko = ( 1.02,
0.10, 
0.11, 
0.20,
0.32, 
0.50, 
3.16, 
1.48, 
0,
0.10)

Kharchenko_err_inf = (0.21,
                      0.04,
                      0.05,
                      0.04,
                      0.07,
                      0.10,
                      0.65,
                      0.30,
                      0,
                      0.04)
                      
Kharchenko_err_sup = (0.26,
                      0.08,
                      0.09,
                      0.05,
                      0.08,
                      0.13,
                      0.82,
                      0.38,
                      0,
                      0.08)
```

```python
Dias = ( 0.659,
        0.010,
        0.190,
        0.104,
        0,
        0,
        0,
        1.754,
        0.029,
        0.115)

Dias_err_inf = (0.536,
                0.003,
                0.085,
                0.087,
                0,
                0,
                0,
                0.275,
                0.017,
                0.042)
                
Dias_err_sup = (2.865,
                0.004,
                0.154,
                0.512,
                0,
                0,
                0,
                0.326,
                0.041,
                0.066)
```

```python
CG= ( 0.005,
     0.105,
     0,
     0.045,
     0,
     0,
     0,
     1.413,
     0.214,
     0.093)

CG_err_inf = (0.002,
              0.039,
              0,
              0.016,
              0,
              0,
              0,
              0.413,
              0.062,
              0.034)

              
CG_err_sup = (0.003,
              0.061,
              0,
              0.026,
              0,
              0,
              0,
              0.583,
              0.088,
              0.055)
```

```python
plt.figure(figsize=(15,10))
plt.style.use('dark_background')

plt.subplot(2,1,1)

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha= 0.1)

plt.scatter(labels, Garcia, marker='*', c='cornflowerblue', s = 500, alpha=0.4, label='Este trabajo')
plt.errorbar(labels, Garcia, yerr=(Garcia_err_inf, Garcia_err_sup), fmt='.', c='cornflowerblue', capsize=5, alpha=0.8)

x = np.arange(0, 10)

for i in range(len(labels)):
    
    if Kharchenko[i] != 0:
        plt.scatter(x[i]-0.15, Kharchenko[i], marker='s', c='gold', s = 200, alpha=0.4, label='Kharchenko')
        plt.errorbar(x[i]-0.15, Kharchenko[i], yerr=[[Kharchenko_err_inf[i]], [Kharchenko_err_sup[i]]], fmt='.', c='gold', capsize=5, alpha=0.8)
        
    if Dias[i] != 0:
        plt.scatter(x[i]+0.1, Dias[i], marker='^', c='tomato', s = 200, alpha=0.4)
        plt.errorbar(x[i]+0.1, Dias[i], yerr=[[Dias_err_inf[i]], [Dias_err_sup[i]]], fmt='.', c='tomato', capsize=5, alpha=0.8)

    if CG[i] != 0:
        plt.scatter(x[i]-0.1, CG[i], marker='X', c='springgreen', s = 200, alpha=0.4)
        plt.errorbar(x[i]-0.1, CG[i], yerr=[[CG_err_inf[i]], [CG_err_sup[i]]], fmt='.', c='springgreen', capsize=5, alpha=0.8)


plt.ylabel('Edad [Gyr]')
plt.legend(['Este trabajo', 'Kharchenko', 'Dias', 'CG'], markerscale=0.7, ncols=2)





plt.subplot(2,1,2)

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha= 0.1)

plt.scatter(labels, Garcia, marker='*', c='cornflowerblue', s = 500, alpha=0.4, label='Este trabajo')
plt.errorbar(labels, Garcia, yerr=(Garcia_err_inf, Garcia_err_sup), fmt='.', c='cornflowerblue', capsize=5, alpha=0.8)

x = np.arange(0, 10)

for i in range(len(labels)):
    
    if Kharchenko[i] != 0:
        plt.scatter(x[i]-0.15, Kharchenko[i], marker='s', c='gold', s = 200, alpha=0.4, label='Kharchenko')
        plt.errorbar(x[i]-0.15, Kharchenko[i], yerr=[[Kharchenko_err_inf[i]], [Kharchenko_err_sup[i]]], fmt='.', c='gold', capsize=5, alpha=0.8)
        
    if Dias[i] != 0:
        plt.scatter(x[i]+0.1, Dias[i], marker='^', c='tomato', s = 200, alpha=0.4)
        plt.errorbar(x[i]+0.1, Dias[i], yerr=[[Dias_err_inf[i]], [Dias_err_sup[i]]], fmt='.', c='tomato', capsize=5, alpha=0.8)

    if CG[i] != 0:
        plt.scatter(x[i]-0.1, CG[i], marker='X', c='springgreen', s = 200, alpha=0.4)
        plt.errorbar(x[i]-0.1, CG[i], yerr=[[CG_err_inf[i]], [CG_err_sup[i]]], fmt='.', c='springgreen', capsize=5, alpha=0.8)


plt.ylabel('Edad [Gyr]')
plt.legend(['Este trabajo', 'Kharchenko', 'Dias', 'CG'], markerscale=0.7, ncols=2)

#revisar eje en logaritmo, introducir valores en log
plt.yscale('log')
```
