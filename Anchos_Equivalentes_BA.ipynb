```python
def mi_funcion(tabla, medicion):
    
    distancia = []
    for i in range(len(tabla)):
        distancia.append( np.abs(tabla[i] - medicion) )
        
    minimo = min(distancia)
    posicion = distancia.index(minimo)
    print('La edad más probable es ', edad[posicion], 'Gyr')
    

    distancia.remove(minimo)

    edades = list(edad)
    edades.remove(edad[posicion])
    
    minimo2 = min(distancia)
    posicion2 = distancia.index(minimo2)
    print('La segunda edad probable es ', edades[posicion2], 'Gyr')

    
    return(edad[posicion])


#se me ocurre que podria directamente ordenar las distancias y asi dar las edaddes según el orden de probabilidad por la "distancia"
    
```
