```python
#pasaje de entre (0 ; 360) a entre (-180 ; 180)
def interv(xx):
    a = []
    for i in range(len(xx)):
        if xx[i] > 180:
            a.append(xx[i] - 360)
        else:
            a.append(xx[i])
    return a
```
