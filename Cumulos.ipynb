```python
#pasaje asencion recta entre -180 y 180
def interv(xx):
    a = []
    for i in range(len(xx)):
        if xx[i] > 180:
            a.append(xx[i] - 360)
        else:
            a.append(xx[i])
    return a
```

```python
labels = ('Alessi 18',
'Archinal 1',
'Berkeley 82',
'ESO 312-04',
'ESO 430-09',
'ESO 435-48',
'ESO 445-74',
'ESO 518-SC03',
'BH 92',
'NGC 6204')
```

```python
plt.figure(figsize=(15,10))
plt.style.use('dark_background')
#plt.style.use('Solarize_Light2')
#plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')

plt.xlim(180,-180)
plt.ylim(-90,90)
plt.axhline(0, ls='--', c = 'grey')
plt.axvline(0, ls='--', c = 'grey')
plt.text(0,2, 'Centro', c='gold', fontsize=10, ha='center', va='center')
plt.text(0,-2, 'galctico', c='gold', fontsize=10, ha='center', va='center')

plt.axhline(30, ls='--', lw=1, c = 'gold')
plt.axhline(-30, ls='--', lw=1, c = 'gold')
plt.text(170, 30, '30', color='gold', fontsize=12, ha='center', va='bottom')
plt.text(170, -30, '-30', color='gold', fontsize=12, ha='center', va='bottom')

plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.scatter(interv(long), lat, marker='*', c='cornflowerblue', s = 500, alpha=0.4)

for i in range(len(interv(long))):
    if labels[i] == 'Alessi 18':
        plt.text(interv(long)[i] - 5, lat[i] - 6, labels[i], color='white', fontsize=12, ha='center', va='top')
    elif labels[i] == 'Archinal 1':
        plt.text(interv(long)[i] - 3, lat[i] - 5, labels[i], color='white', fontsize=12, ha='center', va='top')
    elif labels[i] == 'Berkeley 82':
        plt.text(interv(long)[i] + 3, lat[i] + 5, labels[i], color='white', fontsize=12, ha='center', va='bottom')
    elif labels[i] == 'ESO 312-04':
        plt.text(interv(long)[i] + 5, lat[i] - 5, labels[i], color='white', fontsize=12, ha='center', va='top')
    else:
        plt.text(interv(long)[i], lat[i] + 5, labels[i], color='white', fontsize=12, ha='center', va='bottom')

plt.show()

```
